```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
```

```python
m
```

```python
c = 3 * 10**8# Скорость света, метров в секунду
```

```python
E = m*c**2
```

```python
E
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)
plt.show()
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title("E=mc^2")
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.xlim(0,10)

# LOG SCALE
plt.yscale("log")
plt.grid(which='both',axis='y')

plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=5)
plt.grid(which='both',axis='y')
```
